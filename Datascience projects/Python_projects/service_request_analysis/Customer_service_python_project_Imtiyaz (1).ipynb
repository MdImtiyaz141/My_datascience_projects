# Hypothesis Testing
## time response across all the complaints is similar or not

```python
from scipy import stats
from scipy.stats import ttest_1samp
```

```python
clt = data[data['Complaint Type']=='Noise - Commercial']['Request_close_time']
```

```python
# We have to compare more then two group so we will perform the ANOVA test.
# As we saw above data is skewed so we first appli log distribution then anova test.
```

```python
data['Request_close_time'] = data['Request_close_time'].astype('timedelta64[m]')+1
```

```python
d = {}
for i in data['Complaint Type'].unique():
    d[i] = np.log(data[data['Complaint Type']==i]['Request_close_time'])
```

```python
d['Noise - Commercial'].hist(bins=100)
```

```python
# data is normal distributed so now we will perform the hypotesis testing
```

```python
d.keys()
```

```python
from scipy.stats import f_oneway
stat, p = f_oneway(d['Noise - Street/Sidewalk'], d['Blocked Driveway'], d['Illegal Parking'],
                   d['Derelict Vehicle'], d['Noise - Commercial'], d['Noise - House of Worship'],
                   d['Posting Advertisement'], d['Noise - Vehicle'], d['Animal Abuse'])
print(stat,p)
alpha=0.05

if p>alpha:
    print('distribution is same, hence hypothesis is accpedted')
else:
    print('distribution is not same, hypothesis is rejected')
```
