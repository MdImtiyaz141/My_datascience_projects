```python
# NAN values in categorical variables
features_nan = [feature  for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes=='O'] 
for feature in features_nan:
    print('{}: {}% missing values'.format(feature,np.round(data[feature].isnull().mean(),4)))
```

```python
# replacing missing values with a new entry
def rep_cat_feature(data,features_nan):
    df = data.copy()
    df[features_nan] = df[features_nan].fillna('Missing')
    return df
data = rep_cat_feature(data,features_nan)
data[features_nan].isnull().sum()
```

```python
numerical_variables_nan = [features for features in data.columns if data[features].isnull().sum()>1 and data[features].dtypes!='O']
for feature in numerical_variables_nan:
    print('{}: {}% missing values'.format(feature,np.round(data[feature].isnull().mean(),4)))
```

```python
for feature in numerical_variables_nan:
    median_val = data[feature].median()
    data[feature+'nan'] = np.where(data[feature].isnull(),1,0)
    data[feature].fillna(median_val,inplace=True)
    
data[numerical_variables_nan].isnull().sum()
```

```python
def find_correlation(df,corr_limit=0.60):
    high_correlations = []
    low_correlations = []
    for col in numerical_variables:
        if col == 'SalePrice':
            pass
        else:
            correlation = df[[col,'SalePrice']].corr().loc[[col,'SalePrice']]
            print(col,correlation)
            if (abs(correlation).any() > corr_limit).any():
                high_correlations.append(col+': '+str(correlation))
            else:
                low_correlations.appned(col+': '+str(correlation))
    return low_correlations,high_correlations
```

```python
low_corr,high_corr = find_correlation(data)
```

```python
def one_hot_encoder(df, categorical_feature,nan_as_category=True):
    original_cols = list(df.columns)
    df = pd.get_dummies(df,columns=categorical_feature,dummy_na=nan_as_category,drop_first=True)
    new_columns=[c for c in df.columns if c not in original_cols]
    return df,new_columns
```

```python
data,new_cols_ohe = one_hot_encoder(data,categorical_feature)
```
